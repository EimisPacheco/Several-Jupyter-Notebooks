```python
#let's create a list to extract the number of available Data Scientist jobs at any given time 
#and the Data Scientist jobs in the last 24 hours 
number_jobs_list = soup.title.text.split()
print(number_jobs_list)
print('Total Data Scientist jobs: ' + number_jobs_list[0])
print('Total Data Scientist jobs in the last 24 hours: ' + number_jobs_list[7].lstrip('('))
```

```python
#I use the set to eliminate some possible repeated links
href_set = set()
count=0
for link in soup.findAll('a', attrs={'href': re.compile("^https://")}):
    print (link.get('href').split('?')[0])
    href_set.add(link.get('href').split('?')[0])  
```

```python
href_set = set()
count=0
#The previous pattern is so generic, so it brings links that we do not need
#We need only those that say / jobs / view /, so we must filter by this other pattern
for link in soup.findAll('a', attrs={'href': re.compile("^https://www.linkedin.com/jobs/view/")}):
    #print (link.get('href').split('?')[0])
    href_set.add(link.get('href').split('?')[0]) 
#We count and show all the collected links, linkedin just allows you to see up to 25 
#job links per page
for href in href_set:
    count=count+1
    print(str(count) + '\n')
    print(href)  
```

```python
#Convert to a list to be able to refer to it by an index
href_set = list(href_set)
#Let's use a get request to see more information about the links that we have collected
req = requests.get(href_set[0], data=login_data, headers=headers)
#We are storing the content of the webpage that we took from the list
soup = BeautifulSoup(req.content, 'html.parser')
#in the title we can see really usefull information such as the hiring company,
#the job title, the City, the state and the country
print (soup.title.text)
```

```python
job_info_list = soup.title.text.split(',')
job_info_list[0].split(' hiring ')
```

```python
#By these transformations we are getting the company name, job name and some address
#information of the job, lists are useful for this task
job_info_list = soup.title.text.replace(',','')
print(job_info_list)
trans_job_info_list = [job_info_list.split(' hiring ')[0], 
                       job_info_list.split(' hiring ')[1].split(' in ')[0], 
                       job_info_list.split(' hiring ')[1].split(' in ')[1].split('|')[0].strip()]
#We print the transformed list
print(trans_job_info_list)
```

```python
#We can see all the information regarding the job that we took from the list
soup.text
```

```python
#Now that we undestand how this works, let apply the same above logic to all our list
#So we can stored the recollected information in a csv file with pandas
trans_job_info_list = list()
for i in range(0,len(href_set)):
    #Convert to a list to be able to refer to it by an index
    href_set = list(href_set)
    #Let's use a get request to see more information about the links that we have collected
    req = requests.get(href_set[i], data=login_data, headers=headers)
    #We are storing the content of the webpage that we took from the list
    soup = BeautifulSoup(req.content, 'html5lib')
    #in the title we can see really usefull information such as the hiring company,
    #the job title, the City, the state and the country
    print (soup.title.text)
    
    #By these transformations we are getting the company name, job name and some address
    #information of the job, lists are useful for this task
    job_info_list = soup.title.text.replace(',','')
    trans_job_info_list.append([job_info_list.split(' hiring ')[0], 
                    job_info_list.split(' hiring ')[1].split(' in ')[0], 
                    job_info_list.split(' hiring ')[1].split(' in ')[1].split('|')[0].strip()])
    #We print the transformed list
    print(trans_job_info_list[i])
    print('\n')
```

```python
#Creation a dataframe with the recolected job information
data_scientist_job_df = pd.DataFrame(trans_job_info_list, columns=('Company', 'Job Name', 'Job Address'))
data_scientist_job_df
```

```python
#We are storing this information in a csv file
data_scientist_job_df.to_csv('Linkedin_data_scientist_job.csv')
```
