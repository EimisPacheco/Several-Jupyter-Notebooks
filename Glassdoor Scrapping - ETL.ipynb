```python
soup = BeautifulSoup(req.content, 'html5lib')
print (soup.title.text)
```

```python
#I use the set to eliminate some possible repeated links
href_set = set()
count=0
#We need only those that say /partner/jobListing.htm, so we must filter by this  pattern
for link in soup.findAll('a', attrs={'href': re.compile("^/partner/jobListing.htm")}):
    #We concat https://www.glassdoor.com to be able to click on the link
    #and then open these links later to check them more thoroughly

    #We put it in a list to avoid repeated
    href_set.add('https://www.glassdoor.com'+link.get('href'))  
    
for href in href_set:
    count=count+1
    print(str(count) + '\n')
    print(href)  
```

```python
#Convert the set to a list to be able to refer to it by an index
href_set = list(href_set)
#Let's use a get request to see more information about the links that we have collected
req = requests.get(href_set[5], data=login_data, headers=headers)
#We are storing the content of the webpage that we took from the list
soup = BeautifulSoup(req.content, 'html5lib')
#in the title we can see really usefull information such as the hiring company,
#the job title, the City and the state
print (soup.title.text)
```

```python
#By these transformations we are getting the company name, job name and some address
#information of the job, lists are useful for this task
job_info_list = soup.title.text.split(' Job in ')
print(job_info_list)
print(job_info_list[1].split(' | ')[0])

trans_job_info_list = [job_info_list[0],  
                       job_info_list[1].split(' | ')[0]]
#We print the transformed list
print(trans_job_info_list)
```

```python
#Now that we undestand how this works, let apply the same above logic to all our list
#So we can stored the recollected information in a csv file with pandas
trans_job_info_list = list()
for i in range(0,len(href_set)):
    #Convert to a list to be able to refer to it by an index
    href_set = list(href_set)
    #Let's use a get request to see more information about the links that we have collected
    req = requests.get(href_set[i], data=login_data, headers=headers)
    #We are storing the content of the webpage that we took from the list
    soup = BeautifulSoup(req.content, 'html.parser')

    #By these transformations we are getting the company name, job name and some address
    #information of the job, lists are useful for this task
    job_info_list = soup.title.text.split(' Job in ')
    print(job_info_list)

    trans_job_info_list.append([job_info_list[0], 
                               job_info_list[1].split(' | ')[0]])
    #We print the transformed list
    print(trans_job_info_list[i])
    print('\n')
```

```python
#Creation a dataframe with the recolected job information
data_scientist_job_df = pd.DataFrame(trans_job_info_list, columns=('Company and Job Name', 'Job Address'))
data_scientist_job_df
```
