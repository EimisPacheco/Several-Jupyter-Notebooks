```python
import os
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
pd.options.display.max_rows = 10
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from pandas import DataFrame
from matplotlib import colors as mcolors
```

## Breast Cancer Prediction

```python
Breast_Cancer = pd.read_csv("Breast_Cancer.csv",delimiter=';',decimal=".")
Breast_Cancer.head()
```

**Selecting the Predictor variable and Dependent variable**

```python
X = Breast_Cancer.iloc[:,1:31] 
```

```python
Y = Breast_Cancer.iloc[:,0:1] 
print(Y.head())
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.75, random_state=0)
```

```python
def general_indices(MC, names = None):
    global_precision = np.sum(MC.diagonal()) / np.sum(MC)
    global_error = 1 - global_precision
    precision_category  = pd.DataFrame(MC.diagonal()/np.sum(MC,axis = 1)).T
    positive_precision = (MC[1][1]/(MC[1][0] + MC[1][1])) * 100  #VP/(FN+VP)
    negative_precision= (MC[0][0]/(MC[0][0] + MC[0][1])) * 100 #VN/(VN+FP)
    falses_positives =   MC[0][1]/(MC[0][0] + MC[0][1])         #FP/(VN+FP)
    falses_negatives =   MC[1][0]/(MC[1][0] + MC[1][1])         #FN/(FN+VP)
    positive_assertiveness = MC[1][1]/(MC[0][1] + MC[1][1])       #VP/(FP+VP)
    negative_assertiveness = MC[0][0]/(MC[0][0] + MC[1][0])       #VN/(VN+FN)
    if names!=None:
        precision_category.columns = names
    return {"Confusion Matriz":MC, 
            "Global Precisión":global_precision, 
            "Global Error":global_error, 
            "precision category":precision_category,
            "Positive Precision":positive_precision,
            "Negative Precision":negative_precision,
            "Falses Positives":falses_positives,
            "Falsese Negatives":falses_negatives,
            "Positive Assertiveness":positive_assertiveness,
            "Negative Assertiveness":negative_assertiveness
           }
```

Using the Bayes, Linear Discriminant, and Quadratic Discriminant methods, generate predictive models for the learning table.

```python
#Bayes
# Default Parameters
bayes = GaussianNB()
print(bayes)
#Training the model
bayes.fit(X_train, Y_train)
prediccion = bayes.predict(X_test)

MC = confusion_matrix(Y_test, prediccion)
indices = general_indices(MC,list(np.unique(Y)))
num_indices_list_bayes=[]
nom_indices_list_bayes=[]
for k in indices:
    print("\n%s:\n%s"%(k,str(indices[k])))
    num_indices_list_bayes.append(str(indices[k]))
    nom_indices_list_bayes.append(k)

print("The predictions using naive bayes are: {}".format(bayes.predict(X_test)))
```

```python
#linear discriminant methods
# Default parameters
lda = LinearDiscriminantAnalysis(solver = 'lsqr', shrinkage = 'auto')
print(lda)
lda.fit(X_train, Y_train)
prediccion = lda.predict(X_test)

MC = confusion_matrix(Y_test, prediccion)
indices = general_indices(MC,list(np.unique(Y)))
num_indices_list_dis_lin=[]
nom_indices_list_dis_lin=[]
for k in indices:
    print("\n%s:\n%s"%(k,str(indices[k])))
    num_indices_list_dis_lin.append(str(indices[k]))
    nom_indices_list_dis_lin.append(k)
    
print("Predictions in Testing using linear discriminant are:{}".format(lda.predict(X_test)))
```

```python
#Default parameters
#square discriminant methods
qda = QuadraticDiscriminantAnalysis()
print(qda)
qda.fit(X_train, Y_train)
prediction = qda.predict(X_test)

MC = confusion_matrix(Y_test, prediction)
indices = general_indices(MC,list(np.unique(Y)))
num_indices_list_dis_cua=[]
nom_indices_list_dis_cua=[]
for k in indices:
    print("\n%s:\n%s"%(k,str(indices[k])))
    num_indices_list_dis_cua.append(str(indices[k]))
    nom_indices_list_dis_cua.append(k)
print("Predictions in Testing using quadratic discriminant are: {}".format(qda.predict(X_test)))
```

Using the test table, calculate the confusion matrix, the precision, the positive precision, the negative precision, the false positives, the false negatives, the positive correctness and the negative correctness. Then build a comparison chart.

```python
comparative_frame_df  = pd.DataFrame({'Metric Name':nom_indices_list_bayes, 'Bayes':num_indices_list_bayes,
                                       'Lineal discriminant':num_indices_list_dis_lin,  'Square discriminant':num_indices_list_dis_cua})
comparative_frame_df.reset_index(inplace = True)
comparative_frame_df.set_index('Metric Name', inplace = True)
del(comparative_frame_df['index'])
comparative_frame_df # Comparison chart
```

Construya un cuadro comparativo con respecto a la tarea anterior. ¿Cual metodo es mejor?

R= los resultados aqui presentados cson mejores que la tarea anterior, ya que en general aqui se obtuvo mejor precision global y la precicion negativa fue superior en todos los casos que respecto a la tarea anterior que se tuvo una precision global 96% y negativa del 100% (en el caso del modelo con MLPClassifier solver=lbfgs y  que el modelo MLPClassifier solver=sgd no dio buenos resultados)
