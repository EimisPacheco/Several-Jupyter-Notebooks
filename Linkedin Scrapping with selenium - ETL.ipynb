```python
#function that collects all job links
def collect_all_job_links(href_set):
    count = 0
    href_set_int = set()
    #We are getting the html code from the current page
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    #We need only those that say / jobs / view /, so we must filter by this pattern
    for link in soup.findAll('a', attrs={'href': re.compile("^/jobs/view/")}):
        #print ('https://www.linkedin.com'+link.get('href').split('?')[0])
        #count=count+1
        href_set.add('https://www.linkedin.com'+link.get('href').split('?')[0]) 
        href_set_int.add('https://www.linkedin.com'+link.get('href').split('?')[0]) 
        
    for href in href_set_int:
        count=count+1
        print(str(count) + '\n')
        print(href)
        
    return href_set
```

```python
#This function creates all the list of jobs that is being collected
def collect_all_jobs_head_info(job_list):
    #sentences in which we are not insterested in
    excluded_sentences = ['Be an early applicant', 'Easy Apply', 'connection works here', 'connections work here', 'company alumni work here', 'alumni work here', 'alum works here']
                                                                                                                            
    count = 0
    excluded = None
    job_list_int = list()
    #we can split the result retrieved by jobs_search_panel 
    for elem in jobs_search_panel.text.split('\n'):
        #When elem == 'Create alert for this search' means that we have read all 
        #the jobs result in this jobs_search_pane
        if (elem == 'Create alert for this search') or (elem.isnumeric() == True):
            break
        for i in range(0,len(excluded_sentences)):
            if excluded_sentences[i] in elem:
                excluded = True
                break
            else:
                excluded = False   
        if excluded == False:
            print (elem)
            job_list.append(elem) 
    return job_list
```

```python
#This is to build a list of list in order to create a dataframe with 4 columns from this list
job_list_of_list = []
for i in range(0, len(job_list), 4):
    print(job_list[i:i+4])
    job_list_of_list.append(job_list[i:i+4])
```

```python
#Let's create our datafarme with the previous list
job_list_df = pd.DataFrame(job_list_of_list, columns=( 'Job Name','Company', 'Job Address', 'Posted Time'))
job_list_df.head()
```

```python
#Let's clean the resulted dataframe a little bit getting rid of words that we don't want
job_list_df['Job Name'] = job_list_df['Job Name'].str.rstrip('Promoted')
job_list_df['Posted Time'] = job_list_df['Posted Time'].str.rstrip('ago')
job_list_df.head()
```

```python
list(job_list_df['Job Address'].str.split(','))
```

```python
city_list = []
state_list = []
#We are going to go through a list of list
for i in range(0, len(job_list_df['Job Address'])):
    city_list.append(job_list_df['Job Address'].str.split(',')[i][0])
    
    try:
        #this try statement is because sometimes we are not going to have other data,
        #just the city
        if 'D.C.' in job_list_df['Job Address'].str.split(',')[i][1]:
            state_list.append('DC')
        else:
            state_list.append(job_list_df['Job Address'].str.split(',')[i][1])

    except:
        #in this case we append a None value, or other kind of treatment
        #If city is san francisco then state is california
        if 'San Francisco' in job_list_df['Job Address'].str.split(',')[i][0]:
            state_list.append('CA')
         #If city is san Washington then state is DC
        elif 'Washington' in job_list_df['Job Address'].str.split(',')[i][0]:
            state_list.append('DC')
        else:
            state_list.append(None)
```

```python
state_list
```

```python
job_split_address_dict = {'City':city_list, 'State':state_list}
job_split_address_df = pd.DataFrame(job_split_address_dict)
job_split_address_df.State = job_split_address_df.State.str.lstrip(' ')
job_split_address_df.head()
```

```python
#we have removed the blank spaces
job_split_address_df.State.values
```

```python
#Let's now create another dataframe to get the address details in each column
#this did not work for every case, as the Job Address did not have the same format
job_split_address_list = list(job_list_df['Job Address'].str.split(','))
job_split_address_df = pd.DataFrame(job_split_address_list, columns=( 'City', 'State', 'Country'))
job_split_address_df.head()
```

```python
#Our last step is to concatenate the iniciatl dataframe with the address detail dataframe
#This would be our final DataFrame
aux_job_list = [job_list_df,job_split_address_df]
linkedin_job_list_df = pd.concat(aux_job_list, axis=1)
linkedin_job_list_df.head()
```

```python
#We are storing this information in a csv file
linkedin_job_list_df.to_csv('Linkedin_data_scientist_job.csv')
```
